```python
df = df[df['contact_email'].str.contains('@')]
```

```python
#define function to get domain name from string
def get_domain (y):
    """Takes domain name string field """
    #reg = re.compile('\d{5}')
    #z = reg.findall(y)
    z = y.split('@')
    
    if len(z) == 1:
        return None
    else:
        return z[1] 
```

```python
df['domain'] = df['contact_email'].apply(get_domain)
```
